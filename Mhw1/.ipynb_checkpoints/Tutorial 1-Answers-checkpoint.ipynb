### Multivariate function

I want to know what function $f(x,y)=x^4-x^2+y^2-2xy-2$ looks like. How to do this?
* Use 3d plot
* Use contour

```python
# 3d plot
# To plot 3D graphs, you need to use 3D axes
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

a=np.linspace(-2,2,100)
x,y=np.meshgrid(a,a)
```

```python
fig=plt.figure()
ax=fig.gca(projection='3d') # gca stands for get current axis
ax.plot_surface(x,y,f)
ax.view_init(20,30)
```

```python
# Contours
fig,ax=plt.subplots()
myContour=ax.contour(x,y,f,50)
fig.colorbar(myContour)
```

## Golden section
Given the following function, can you find its minimum value in range $[0,9]$ using golden section?

```python
def func(x):
    isarray=type(x) is np.ndarray
    coefs=np.array([8.001371477,
    -24.06731415,
    37.076044,
    0,
    -43.86909846,
    44.42701101,
    -22.0126204,
    6.536434989,
    -1.248082478,
    0.157159012,
    -0.012990941,
    0.000678657,
    -2.03269E-05,
    2.66065E-07
    ])
    base=np.zeros((len(x) if isarray else 1,14))
    for i in range(1,14):
        base[:,i]+=x**i
    result=base.dot(coefs)
    return result if isarray else sum(result)

x=np.linspace(0,9,200)
y=func(x)
plt.plot(x,y);
```

```python
from numpy.linalg import norm
def golden_section(func,start,end,reference,tol):
    if norm(start - end) < tol:
        # return the final point and the evaluation when the start and the end points are within range of convergence
        return {"x":reference, "evaluation":func(reference)}
    else:
        # evaluate function values at start point, end point and reference point
        y_start=func(start)
        y_end=func(end)
        y_reference=func(reference)
        if norm(reference - start) > norm(end - reference):
            # reference point is at the right side of the middle point 
            # decide new reference point, and do function evaluation
            new_reference=start+(reference-start)*0.618
            y_new_reference=func(new_reference)
                
            if y_new_reference < y_reference:
                # recursively call this golden section function, select new start, end and ref 
                return golden_section(func,start,reference,new_reference,tol)
            else:
                # recursively call this golden section function, select new start, end and ref 
                return golden_section(func,new_reference,end,reference,tol)
        else:
            # reference point is at the left side of the middle point 
            new_reference=end-(end-reference)*0.618
            y_new_reference=func(new_reference)
            if y_new_reference<y_reference:
                return golden_section(func,reference,end,new_reference,tol)
            else:
                return golden_section(func,start,new_reference,reference,tol)
```

```python
golden_section(func,0,9,9*0.618,1e-5)
```

## Steepest descent
Let's also try steepest descent on the same function!

```python
def func_der(x):
    isarray=type(x) is np.ndarray
    coefs=np.array([8.001371477,
    -24.06731415,
    37.076044,
    0,
    -43.86909846,
    44.42701101,
    -22.0126204,
    6.536434989,
    -1.248082478,
    0.157159012,
    -0.012990941,
    0.000678657,
    -2.03269E-05,
    2.66065E-07
    ])
    coefs_der = np.arange(14) * coefs
    base=np.zeros((len(x) if isarray else 1,14))
    for i in range(1,14):
        base[:,i]+=x**(i-1)
    result=base.dot(coefs_der)
    return result if isarray else sum(result)
```

```python
def steepest_descent(func,first_derivate,starting_point,stepsize,tol):
    # evaluate the gradient at starting point
    deriv=first_derivate(starting_point)
    count=0
    while norm(deriv) > tol and count < 1e6:
        # calculate new point position
        new_point = starting_point - stepsize * deriv
        if func(new_point) < func(starting_point):
            # the step makes function evaluation lower - it is a good step. what do you do?
            starting_point = new_point
            deriv = first_derivate(starting_point)
            stepsize *= 1.2
            
        else:
            # the step makes function evaluation higher - it is a bad step. what do you do?
            stepsize /= 2
        count+=1
    # return the results
    return {"x":starting_point, "evaluation":func(starting_point), "count":count}
```

```python
steepest_descent(func, func_der, 6, 0.1, 1e-5)
```

## Minimizing a function with scipy
The multivariate Rosenbrock function is given below:
$f(\vec{x})=\sum_{i=1}^{N-1}100(x_{i+1}-x_i^2)^2+(1-x_i^2)^2$

How to minimize the function in 5d?

```python
from scipy import optimize
```

```python
def func(X):
    f=sum(100*(X[1:]-X[:-1]**2)**2+(1-X[:-1]**2)**2) # matrix writing
    return f

x0=np.array([1.3,0.7,0.8,1.9,1.2])
res=optimize.minimize(func,x0,method='BFGS',options={"gtol":1e-5,"disp":True})
```

More on optimization: [Scipy tutorial on optimization](https://docs.scipy.org/doc/scipy/reference/tutorial/optimize.html)
